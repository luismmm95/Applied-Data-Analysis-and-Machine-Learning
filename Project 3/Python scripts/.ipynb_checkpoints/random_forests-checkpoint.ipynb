# Project 3 - Random forests

```python
#Loading data and preprocesiing
import pandas as pd
import numpy as np
import Neural as nn
import IO
import Preprocess
import time

raw_data = IO.loader("hmnist_28_28_RGB.csv")

data, raw_labels = Preprocess.create_label_data_split(raw_data, "label")

labels, legend = Preprocess.make_categorical(raw_labels, 7)

data = data.astype(np.float64)

train, train_labels, test, test_labels = Preprocess.test_train_split(data, raw_labels)
```

## Evaluating the Random forest model with different values of hyperparameter using ScikitLearn

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [100,400],
                          'max_leaf_nodes': [100,200,300]}
model=RandomForestClassifier(random_state=0,n_jobs=-1)
grid = GridSearchCV(model, param_grid, cv=5)

grid.fit(train, train_labels)
```

```python
cvres = grid.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(mean_score, params)
```

```python
for mean_score, params in zip(cvres["mean_train_score"], cvres["params"]):
    print(mean_score, params)
```

```python
print(grid.best_params_)
```

## Evaluating the best found model 

```python
#from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=400,max_leaf_nodes=300, random_state=0,n_jobs=-1)
model.fit(train, train_labels)

ypred_test = model.predict(test)
```

```python
#Printing results for the train data
from sklearn import metrics
ypred_train = model.predict(train)
print(metrics.classification_report(train_labels,ypred_train ))
```

```python
#Printing results for the test data
print(metrics.classification_report(test_labels,ypred_test ))
```
